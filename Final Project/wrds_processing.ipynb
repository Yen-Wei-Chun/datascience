# 文字處理

為計算情緒分數，我們需要先將新聞內文做切割，並去除不重要的字詞

```python
import pickle
import pandas as pd

with open('Reuters_news.pickle', 'rb') as file:
    news_data = pickle.load(file)

news_datafr = pd.DataFrame(news_data)[['date', 'title', 'summary', 'content']]
news_datafr.head()
```

```python
with open('money_news_trans_all.pickle', 'rb') as file1:
    money_raw = pickle.load(file1)
```

```python
with open('bbc_tradewar.bin', 'rb') as file2:
    bbc_raw = pickle.load(file2)

bbc_datafr = pd.DataFrame(bbc_raw)[['time', 'title', 'summary', 'doc']]
bbc_datafr.head()
```

# 內文切割

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
from nltk.corpus import stopwords
import re
```

```python
reuters_sentences = []

for i in range(len(news_datafr['content'])):
    sent = sent_tokenize(news_datafr['content'][i].lower()) # 將內文切割成句子
    resent = [re.sub(r'[^\w\s]','', smallsent) for smallsent in sent] # 移除標點符號
    reuters_sentences.append(resent) # sentences: a list consists of many lists(news)
print(len(reuters_sentences))
reuters_sentences[0] # a list
```

```python
bbc_sentences = []

for i in range(len(bbc_datafr['doc'])):
    sent = sent_tokenize(bbc_datafr['doc'][i].lower())
    resent = [re.sub(r'[^\w\s]','', smallsent) for smallsent in sent]
    bbc_sentences.append(resent) # sentences: a list consists of many lists(news)
print(len(bbc_sentences))
bbc_sentences[0] # a list
```

```python
money_sentences = []
for sent in money_raw:
    resent = re.sub(r'[^\w\s]','', sent)
    money_sentences.append(resent.lower())
print(len(money_sentences))
money_sentences[0] # a string
```

```python
reuters_wrds_lst = []
for news_lst in reuters_sentences:
    word_in_sent = [word_tokenize(sent) for sent in news_lst] # 將鋸子處理成單詞
    reuters_wrds_lst.append(word_in_sent)
reuters_wrds_lst[0] #wrds_lst有三層：最外層是大list裝所有新聞，第二層是每一則新聞(共636則)，每一則新聞又含好幾個list裝每句拆成的單詞
```

```python
bbc_wrds_lst = []
for news_lst in bbc_sentences:
    word_in_sent = [word_tokenize(sent) for sent in news_lst]
    bbc_wrds_lst.append(word_in_sent)
bbc_wrds_lst[0] 
```

```python
money_wrds_lst = []
for news_str in money_sentences:
    word_in_sent = word_tokenize(news_str)
    money_wrds_lst.append(word_in_sent)
money_wrds_lst[0] #有兩層，外層裝所有新聞（共3083則），內層裝每則新聞的單詞
```

# 移除停用字

```python
stop_words = list(set(stopwords.words('english')))

print('Number of stopwords:', len(stop_words))
print(f'First 30 stop words:\n{stop_words[:30]}') #格式化表達方法

def remove_stopword(sentence):
    return [w for w in sentence if w not in stop_words] # if word in a sentence is not a stop word, return it
```

```python
filtered_lst = []
for i in range(len(reuters_wrds_lst)):
    news_filtered = []
    for s in reuters_wrds_lst[i]:
        filtered = remove_stopword(s)
        news_filtered.append(filtered)
    filtered_lst.append(news_filtered)
print(len(filtered_lst))
print(filtered_lst[0])
```

```python
bbc_filtered_lst = []
for i in range(len(bbc_wrds_lst)):
    news_filtered = []
    for s in bbc_wrds_lst[i]:
        filtered = remove_stopword(s)
        news_filtered.append(filtered)
    bbc_filtered_lst.append(news_filtered)
print(len(bbc_filtered_lst))
print(bbc_filtered_lst[0])
```

```python
money_filtered_lst = []

for i in range(len(money_wrds_lst)):
    filtered = remove_stopword(money_wrds_lst[i])
    money_filtered_lst.append(filtered)

print(len(money_filtered_lst))
print(money_filtered_lst[0])
```

# 存檔

```python
file = open('RNews_wrds.pickle', 'wb')
pickle.dump(filtered_lst, file)
file.close()
```

```python
file = open('Money_wrds.pickle', 'wb')
pickle.dump(money_filtered_lst, file)
file.close()
```

```python
file = open('BBC_wrds.pickle', 'wb')
pickle.dump(bbc_filtered_lst, file)
file.close()
```
