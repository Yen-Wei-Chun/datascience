# 翻譯

由於台灣經濟日報為中文，為方便後續文字處理，故先進行翻譯。

```python
import pickle
import pandas as pd

with open('money_news.pickle', 'rb') as file1:
    money_news = pickle.load(file1)
    
money_datafr = pd.DataFrame(money_news)[['date', 'title', 'summary']]
money_datafr.head()
```

```python
import jieba
import jieba.analyse
from nltk.corpus import sinica_treebank
import nltk
from googletrans import Translator
```

由於Google無法一次進行大量文件翻譯，所以分次進行。

```python
money_trans_list = []

#cnt = 0
for content in money_datafr['summary'][2901:3084]:
    translator = Translator()
    try:
        trans_to_eng = translator.translate(content, dest = 'en').text
    except:
        print(content)
        continue
    #print(cnt)
    money_trans_list.append(trans_to_eng)
    #cnt += 1
   
```

```python
import pickle
import pandas as pd

with open('money_news_trans 1~2901.pickle', 'rb') as file1:
    money_news_trans = pickle.load(file1)
```

```python
for trans in money_trans_list:
    money_news_trans.append(trans)

len(money_news_trans)
```

```python
import pickle

file = open('money_news_trans_all.pickle', 'wb')
pickle.dump(money_news_trans, file)
file.close()
```
